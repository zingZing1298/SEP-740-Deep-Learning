```python
#importing all necessary libraries for visualization and data processing
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# input variable refers to the directory of the dataset in a local environment
input = "DL Project"
```

## Data Exploration

As discussed in the report the directory has 4 data files(csv files) in the current version of the dataset. 

```python
print(os.listdir("../"+input))
```

### 1st file: ../input/list_attr_celeba.csv


```python
attr = pd.read_csv("../" +input+ "/list_attr_celeba.csv")

# shuffling the image data within the dataframe
samples = attr.sample(frac=1, random_state=42).reset_index(drop=True)
samples.head()
```

```python
fig, ax = plt.subplots(figsize=(30,20), dpi=50)
plot = samples.hist(ax=ax, layout=(5,10),bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
           xlabelsize=8, ylabelsize=8, grid=False)
```

The above histograms provide us with in-depth analysis of the dataset. Some notable deductions are as follows:

<ul>
    <li> There are more female identities in the dataset than male, hence the number of features such as beard, mustache, goatee, reducing hairline, sideburns or other male- related attribute data are low
    <li> Facial attributes common among woman such are wearing earrings, wearing lipstick, etc are more in number
    <li> Most data points are Young individuals

</ul>

```python
a4_dims = (13, 10)
# finding correlation matrix and plotting heatmap of the correlation matrix
corrM = round(samples.corr(),2)
fig, ax = plt.subplots(figsize=a4_dims)
sns.heatmap(ax=ax,data = corrM)
```

The above heat map helps us understands positive and negative correlation in the correlation matrix. A good amount of strong correlation can be observed
